## Addestramento di un modello customizzato di rete neurale convoluzionale con ottimizzazione degli iperparametri con il metodo grid search

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

def create_model(num_classes, kernel_size=3, num_neurons=128):
    class ComplexCNN(nn.Module):
        def __init__(self):
            super(ComplexCNN, self).__init__()
            # Strati convoluzionali
            self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
            self.bn1 = nn.BatchNorm2d(64)
            self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
            self.bn2 = nn.BatchNorm2d(128)
            self.conv3 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
            self.bn3 = nn.BatchNorm2d(256)
            self.conv4 = nn.Conv2d(256, 512, kernel_size=3, padding=1)
            self.bn4 = nn.BatchNorm2d(512)

            # Strato di dropout per la regolarizzazione
            self.dropout = nn.Dropout(0.5)

            # Strati fully connected
            self.fc1 = nn.Linear(512 * 14 * 14, 1024)  # Assumendo input di dimensioni [224, 224]
            self.fc2 = nn.Linear(1024, 512)
            self.fc3 = nn.Linear(512, num_classes)

        def forward(self, x):
            x = self.pool(F.relu(self.bn1(self.conv1(x))))
            x = self.pool(F.relu(self.bn2(self.conv2(x))))
            x = self.pool(F.relu(self.bn3(self.conv3(x))))
            x = self.pool(F.relu(self.bn4(self.conv4(x))))
            x = x.view(x.size(0), -1)  # Preserva la dimensione del batch
            x = F.relu(self.fc1(self.dropout(x)))
            x = F.relu(self.fc2(self.dropout(x)))
            x = self.fc3(x)
            return x

        def pool(self, x):
            return F.max_pool2d(x, 2)

    return ComplexCNN()
```

```python
param_grid = {
    'kernel_size': [3, 5],  # Dimensioni comuni per i kernel delle CNN
    'num_neurons': [64, 128],  # Numero di neuroni nei layer fully connected
    "learning_rate": [0.01, 0.001]
}
```

```python
def train_model(model, train_loader, validation_loader, criterion, optimizer, num_epochs=10):
    best_model_wts = copy.deepcopy(model.state_dict())
    best_accuracy = 0.0

    for epoch in range(num_epochs):
        model.train()  # Imposta il modello in modalitÃ  di training
        running_loss = 0.0
        running_corrects = 0

        # Ciclo di training
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            # Calcola la perdita e il numero di predizioni corrette
            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_loader.dataset)
        epoch_accuracy = running_corrects.double() / len(train_loader.dataset)

        print(f'Epoca {epoch+1}/{num_epochs}:')
        print(f'- Training Loss: {epoch_loss:.4f} Acc: {epoch_accuracy:.4f}')

        # Valutazione sul set di validazione
        model.eval()
        validation_loss = 0.0
        validation_corrects = 0

        with torch.no_grad():
            for inputs, labels in validation_loader:
                inputs, labels = inputs.to(device), labels.to(device)

                outputs = model(inputs)
                loss = criterion(outputs, labels)

                validation_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                validation_corrects += torch.sum(preds == labels.data)

        validation_loss /= len(validation_loader.dataset)
        validation_accuracy = validation_corrects.double() / len(validation_loader.dataset)

        print(f'- Validation Loss: {validation_loss:.4f} Acc: {validation_accuracy:.4f}')

        # Salva il modello se Ã¨ il migliore finora
        if validation_accuracy > best_accuracy:
            best_accuracy = validation_accuracy
            best_model_wts = copy.deepcopy(model.state_dict())

    model.load_state_dict(best_model_wts)
    return model, best_accuracy
```

```python
best_accuracy = 0.0
best_model = None

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Itera sui parametri
for kernel_size in param_grid['kernel_size']:
    for num_neurons in param_grid['num_neurons']:
        for learning_rate in param_grid["learning_rate"]:
            print(f"Training with kernel_size={kernel_size}, num_neurons={num_neurons}, learning_rate={learning_rate}")

            model = create_model(num_classes=num_classes, kernel_size=kernel_size, num_neurons=num_neurons)
            model = model.to(device)

            criterion = nn.CrossEntropyLoss()
            optimizer = optim.Adam(model.parameters(), lr=learning_rate)

            trained_model, accuracy = train_model(model, train_loader_augmented, validation_loader, criterion, optimizer)

            if accuracy > best_accuracy:
                best_accuracy = accuracy
                best_model = copy.deepcopy(trained_model)

# Salva il miglior modello
torch.save(best_model.state_dict(), 'best_model.pth')
print(f"Best model saved with accuracy: {best_accuracy:.4f}")
```

## Addestramento modelli transfer learning

```python
num_epochs=100

# Definisci una funzione per l'allenamento con early stopping
def train_model_with_early_stopping(model, train_loader, validation_loader, optimizer, save_path='best_model.pth', patience=10):
    criterion = nn.CrossEntropyLoss()

    best_accuracy = 0.0
    epochs_without_improvement = 0

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        print(f'Epoch {epoch+1}, Loss: {running_loss/len(train_loader)}')

        # Validazione
        model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, labels in validation_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        accuracy = 100 * correct / total
        print(f'Validation Accuracy: {accuracy}%')

        # Salva il modello se ottieni una miglior accuratezza
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            torch.save(model.state_dict(), save_path)
            print(f'Model saved with accuracy: {best_accuracy}%')
            epochs_without_improvement = 0
        else:
            epochs_without_improvement += 1

        # Check for early stopping
        if epochs_without_improvement >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break
```

#### Si testano tre diversi modelli: resnet50, VGG16 e EfficientNet-B0

```python
# Carica il modello ResNet50 preaddestrato
resnet50 = models.resnet50(pretrained=True)

# Blocca i pesi degli strati convoluzionali
for param in resnet50.parameters():
    param.requires_grad = False

# Modifica l'ultimo livello per adattarlo al numero di classi del tuo dataset
resnet50.fc = nn.Linear(resnet50.fc.in_features, num_classes)

# Sblocca i pesi dell'ultimo livello completamente connesso
for param in resnet50.fc.parameters():
    param.requires_grad = True

resnet50 = resnet50.to(device)

optimizer = optim.Adam(resnet50.fc.parameters(), lr=0.001)

# Allenamento con ResNet50
print("Training ResNet50")
train_model_with_early_stopping(resnet50, train_loader_augmented, validation_loader, optimizer, save_path="/content/drive/MyDrive/Colab_Notebooks/resnet50.pth")
```

```python
# Carica il modello VGG16 preaddestrato
vgg16 = models.vgg16(pretrained=True)

# Blocca i pesi degli strati convoluzionali
for param in vgg16.features.parameters():
    param.requires_grad = False

# Sblocca i pesi di tutti gli strati nel blocco classifier
for param in vgg16.classifier.parameters():
    param.requires_grad = True

# Modifica l'ultimo livello completamente connesso per adattarlo al numero di classi
vgg16.classifier[6] = nn.Linear(vgg16.classifier[6].in_features, num_classes)

vgg16 = vgg16.to(device)

# Solo i parametri del nuovo livello completamente connesso saranno aggiornati
optimizer = torch.optim.Adam(vgg16.classifier[6].parameters(), lr=0.001)

# Allenamento con VGG16
print("Training VGG16")
train_model_with_early_stopping(vgg16, train_loader_augmented, validation_loader, optimizer, save_path="/content/drive/MyDrive/Colab_Notebooks/vgg16.pth")
```

```python
# Carica il modello EfficientNet-B0 preaddestrato
efficientnet_b0 = models.efficientnet_b0(pretrained=True)

# Blocca i pesi degli strati convoluzionali
for param in efficientnet_b0.features.parameters():
    param.requires_grad = False

# Sblocca i pesi di tutti gli strati nel blocco classifier
for param in efficientnet_b0.classifier.parameters():
    param.requires_grad = True

efficientnet_b0.classifier[1] = nn.Linear(efficientnet_b0.classifier[1].in_features, num_classes)

efficientnet_b0 = efficientnet_b0.to(device)

# Configura l'ottimizzatore per aggiornare tutti i parametri del blocco classifier
optimizer = torch.optim.Adam(efficientnet_b0.classifier.parameters(), lr=0.001)

# Allenamento con ResNet50
print("Training EfficientNet-B0")
train_model_with_early_stopping(efficientnet_b0, train_loader_augmented, validation_loader, optimizer, save_path="/content/drive/MyDrive/Colab_Notebooks/efficientnet_b0.pth")
```

#### Il modello con la miglior accuratezza Ã¨ ResNet50. Proviamo a riaddestrarlo applicando tecniche di validation e regolarizzazione per migliorare le performance del modello

```python
resnet50_retrain = models.resnet50(pretrained=True)

for param in resnet50_retrain.parameters():
    param.requires_grad = False

num_features = resnet50_retrain.fc.in_features

# Si aggiunge anche uno strato di dropout al 50% per ridurre la possibilitÃ  di overfitting
resnet50_retrain.fc = nn.Sequential(
    nn.Linear(num_features, 1024),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(1024, num_classes)
)

resnet50_retrain = resnet50_retrain.to(device)
```

```python
# Definizione dell'ottimizzatore con L2 Regularization (Weight Decay)
optimizer = optim.Adam(resnet50_retrain.fc.parameters(), lr=0.001, weight_decay=1e-4)
# definizione di un oggetto "scheduler" che diminuisce il learning rate dell'ottimizzatore ad intervalli regolari durante l'addestramento.
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.1)
criterion = nn.CrossEntropyLoss()

def train_model(model, train_loader, val_loader, optimizer, scheduler, criterion, num_epochs=50, patience=10):
    best_acc = 0.0
    epochs_without_improvement = 0

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct, total = 0, 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_acc = 100 * correct / total
        scheduler.step()

        # Validazione
        model.eval()
        val_correct, val_total = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

        val_acc = 100 * val_correct / val_total
        print(f"Epoch {epoch+1}: Loss {running_loss/len(train_loader):.4f}, Train Acc: {train_acc:.2f}%, Val Acc: {val_acc:.2f}%")

        # Early stopping
        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), "/content/drive/MyDrive/Colab_Notebooks/retrained_model.pth")
            epochs_without_improvement = 0
            print("Model saved!")
        else:
            epochs_without_improvement += 1

        if epochs_without_improvement >= patience:
            print("Early stopping triggered!")
            break

    print("Training complete.")
```

#### Le performance non sono migliorate. Utilizziamo il modello con la migliore accuracy per eseguire il test finale sul test set

```python
import torch
from sklearn.metrics import confusion_matrix, accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Funzione per caricare il modello
def load_model(model, path='best_model.pth'):
    model.load_state_dict(torch.load(path))
    model.eval()  # Imposta il modello in modalitÃ  valutazione
    return model

# Funzione per testare il modello e identificare gli errori
def test_model(model, test_loader, class_names):
    model.eval()
    all_labels = []
    all_predictions = []
    all_images = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            all_labels.extend(labels.cpu().numpy())
            all_predictions.extend(predicted.cpu().numpy())
            all_images.extend(inputs.cpu())  # Salva le immagini originali

    # Calcola l'accuracy
    accuracy = accuracy_score(all_labels, all_predictions)
    print(f'Test Accuracy: {accuracy * 100}%')

    # Calcola la matrice di confusione
    conf_matrix = confusion_matrix(all_labels, all_predictions)
    print('Confusion Matrix:')
    # print(conf_matrix)

    # Visualizza la matrice di confusione
    plt.figure(figsize=(10, 7))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names, cbar=False)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

    # Ottieni gli errori di predizione
    errors = [(img, true, pred) for img, true, pred in zip(all_images, all_labels, all_predictions) if true != pred]

    # Conta gli errori per ogni combinazione di classi
    error_counts = {}
    for _, true, pred in errors:
        key = (true, pred)
        if key not in error_counts:
            error_counts[key] = 0
        error_counts[key] += 1

    # Ordina gli errori per frequenza
    sorted_errors = sorted(error_counts.items(), key=lambda x: x[1], reverse=True)

    # Seleziona le immagini di esempio per gli 8 casi piÃ¹ frequenti
    examples = []
    for (true, pred), _ in sorted_errors[:8]:
        example = next((img for img, t, p in errors if t == true and p == pred), None)
        if example is not None:
            examples.append((example, true, pred))

    return examples
```

```python
# Funzione per visualizzare gli esempi
def show_examples(examples, class_names):
    plt.figure(figsize=(12, 6))
    for i, (img, true, pred) in enumerate(examples):
        plt.subplot(2, 4, i + 1)
        img = img.permute(1, 2, 0)  # Cambia la dimensione da CHW a HWC
        img = img.numpy() * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])  # De-normalizza
        img = np.clip(img, 0, 1)  # Limita i valori tra 0 e 1
        plt.imshow(img)
        plt.title(f'Real: {class_names[true]}\nPred: {class_names[pred]}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

show_examples(error_examples, class_names)
```
